```python
from sklearn.cluster import KMeans
```

```python
KMeans()
```

```python
wcss=[]
for i in range(1,10):
    model=KMeans(init='k-means++',n_clusters=i,random_state=None)
    model.fit(b)
    wcss.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(wcss)
plt.xlabel("Number of clusters")
plt.title("elbow curve")
plt.show()
```

```python
model=KMeans(init='k-means++',n_clusters=3,random_state=None)
model.fit(b)
wcss.append(model.inertia_)
```

```python
pred=model.predict(b)
```

```python
clus_centers=pd.DataFrame(model.cluster_centers_)
clus_centers.columns=b.columns
clus_centers
```

```python
model.inertia_
```

```python
cluster=pd.DataFrame(model.labels_)
cluster.columns=["cluster"]
a1=pd.concat([a,cluster],axis=1)
a1.head()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(pred,model.labels_))
print(confusion_matrix(pred,model.labels_))
```
